```python
import requests
from bs4 import BeautifulSoup
```

```python
BASE_URL = "https://pythonizing.github.io/data/real-estate/rock-springs-wy/LCWYROCKSPRINGS/"

r = requests.get(BASE_URL, verify = False) # verify to False since site doesn't support SSL
soup = BeautifulSoup(r.content, "html.parser")


total_pages = int(soup.find_all("a", {"class": "Page"})[-1].text)
print(total_pages)
BASE_URL = "https://pythonizing.github.io/data/real-estate/rock-springs-wy/LCWYROCKSPRINGS/t=0&s={0}"
```

```python
data_rows_list = []

for page in range(0, total_pages*10, 10):
    url = BASE_URL.format(page)
    r = requests.get(url, verify = False) # verify to False since site doesn't support SSL
    c = r.content
    soup = BeautifulSoup(c, "html.parser")

    all = soup.find_all("div", {"class": "propertyRow"})

    for row in all:
        # find price
        data_dict = {}
        price = row.find("h4", {"class": "propPrice"}).text.strip() # remove empty spaces around text
        # find address
        addresses = row.find_all("span", {"class": "propAddressCollapse"})
        full_address = ''
        for address in addresses:
            full_address += address.text

        # find number of beds
        try:
            bed = row.find("span", {"class": "infoBed"}).find(
                "b").text  # find span tag having this class
        except AttributeError:
            bed = None

        try:
            bath = row.find("span", {"class": "infoValueFullBath"}).find(
                "b").text  # find span tag having this class
        except AttributeError:
            bath = None

        try:
            half_bath = row.find("span", {"class": "infoValueHalfBath"}).find(
                "b").text  # find span tag having this class
        except AttributeError:
            half_bath = None

        try:
            area = row.find("span", {"class": "infoSqFt"}).find(
                "b").text  # find span tag having this class

        except AttributeError:
            area = None

        lotsize = None
        for column_group in row.find_all("div", {"class":"columnGroup"}):
            for feature_group, feature_name in zip(column_group.find_all("span", {"class": "featureGroup"}), column_group.find_all("span", {"class": "featureName"})):
                # Now add code for the properties that we want to extract
                if "Lot Size" in feature_group.text:
                    print("Lotsize: ",feature_name.text)
                    lotsize= feature_name.text
        
        data_dict["Address"]  = full_address
        data_dict["Price"]  = price
        data_dict["Beds"] = bed
        data_dict["Area"] = area
        data_dict["Full Baths"] = bath
        data_dict["Half Baths"] = half_bath
        data_dict["Lot Size"]= lotsize

        data_rows_list.append(data_dict)
        
```
